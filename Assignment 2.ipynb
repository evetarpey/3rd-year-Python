### Introduction: 

The double-slit experiment displays the dual nature of light as both a particle and a wave. Diffraction and interference are important phenomena in this experiment.


When light passes through the slits, it undergoes diffraction. This causes the light waves to spread out- which allows the waves from the two slits to interact and interfere with eachother.

We can then see the waves that interact and interfere with eachother on the screen. When constructive interference occurs, we get regions of higher intensity, which are basically the brighter fringes on the screen! When destructive interference occurs, we get regions of lower intensity, which are shown by the dark fringes on the screen.


### The horizontal variation of the intensity of light at a screen after passing through a double slit:
### $$ I = I_0 (\frac{sin\alpha}{\alpha})^2(cos\beta)^2 $$ ###

where $$ \alpha = \frac{\pi a}{\lambda} sin\theta $$ and $$ \beta = \frac{\pi d}{\lambda} sin\theta $$

- $ I_0 $ is the initial intensity.
- a = 0.09 mm is the width of one of the slits.
- L = 480 mm is the distance from the slits to the screen.
- d = 0.4 mm is the distance between the slits.
- $\lambda$ = 670 nm is the wavelength of the light.
- $\theta$ = angle at which one is observing the experiment from.



### Task 1 :

Plot the predicted 1-D intensity pattern of the normalised intensity of light observed from two slits. 



$$ \frac{I}{I_0} = (\frac{sin\alpha}{\alpha})^2(cos\beta)^2 $$ 
where

- $ \frac{I}{I_0}$ = normalised intensity. 


- $ \alpha $ and $ \beta $ are shown above.

$$ tan\theta = \frac{x}{L} $$


where 
- x = horizontal position on the detector.
- L = distance from the slits to the screen.
- $\theta$ = angle at which one is observing the experiment from.

```python
import numpy as np                                              # import numpy to perform mathematical operations
import matplotlib.pyplot as plt                                 # we must import matplotlib.pyplot as plt to plot our graph 
%matplotlib inline

# use the parameters given (units: mm)
a = 0.09                                                        # the width of one of the slits (mm)
L = 480                                                         # distance from the slits to the screen (mm)
d = 0.4                                                         # the distance between the slits (mm)
w = 670e-6                                                      # the wavelength of light (mm)

# define a function for the double slit intensity pattern
def double_slit_intensity(x):
    theta = np.arctan(x/L)                                      # calculate the angle theta
    alpha = ((np.pi*a)/ w)* np.sin(theta)                       # calculate alpha
    beta = ((np.pi*d)/ w)* np.sin(theta)                        # calculate beta
    
    I_double_slit = (np.sin(alpha)/ alpha)**2 * (np.cos(beta))**2
    return I_double_slit

x = np.linspace(-6, 6, 1000)                                    # horizontal position along the detector (mm)

I_double_slit = double_slit_intensity(x)                        # normalized double slit intensity pattern

plt.figure(figsize=(15, 6))                                     # to enlarge the graph

# Plotting
plt.plot(x, I_double_slit, "m-")                               
plt.title("double slit normalised intensity pattern ")          # to title the graph
plt.xlabel("position on detector (mm)")                         # to label the x axis
plt.ylabel("normalised intensity (I/I_0) ")                     # to label the y axis 
plt.grid()                                                      # to display a grid on the graph
plt.show()                                                      # to show the graph
```

### Task 2: 
On the same figure now plot the predicted 1-D intensity pattern of light observed from just one of the slits.


$$ \frac{I}{I_\theta} = (\frac{sin\alpha}{\alpha})^2 $$

```python
# Define the function for the single slit intensity pattern
def single_slit_intensity(x):
    theta = np.arctan(x/L)                                     # calculate the angle theta
    alpha = ((np.pi*a)/ w)* np.sin(theta)                      # calculate alpha
    beta = ((np.pi*d)/ w)* np.sin(theta)                       # calculate beta
    I_single_slit = (np.sin(alpha) / alpha)**2
    return I_single_slit                                

I_single_slit = single_slit_intensity(x)                       # normalised single slit intensity 

plt.figure(figsize=(15, 6))                                    # to enlarge the graph

# Plot both intensities on one graph
plt.plot(x, I_double_slit, "m-", label='double slit')          # double-slit intensity 
plt.plot(x, I_single_slit, "g-", label='single slit')          # single-slit intensity 

# to title the graph
plt.title("double slit vs. single slit normalised intensity pattern ") 

plt.xlabel("position on detector (mm)")                        # to label the x axis
plt.ylabel("normalised intensity (I/I_0)")                     # to label the y axis 
plt.legend()                                                   # to distinguish between the patterns
plt.grid()                                                     # to display a grid on the graph
plt.show()                                                     # to show the graph
```

### Task 3: Simulate the 2D pattern
Develop your program to simulate the appearance of the pattern in 2-D on the viewing card in the experiment.

- Use an aspect ratio between x and y of 3:2.

```python
y = np.linspace(-6,6,1000)                                      # define an array of vertical positions on the detector

X, Y = np.meshgrid(x, y)                                        # create a 2D grid for x and y using "meshgrid"

# define a function for the double slit intensity pattern using X and Y
def double_slit_intensity(X, Y):
    theta = np.arctan(Y/L)                                      # calculate the angle theta
    alpha = ((np.pi*a)/ w)* np.sin(theta)                       # calculate alpha
    beta = ((np.pi*d)/ w)* np.sin(theta)                        # calculate beta
    I_double_slit = (np.sin(alpha)/ alpha)**2 * (np.cos(beta))**2
    return I_double_slit

Z = double_slit_intensity(X, Y)                                 # calculate the intensity for each point in the grid

# create a viewing card to simulate the appearance of the pattern in 2D
viewing_card = plt.contourf(Y, X, Z, cmap='inferno',levels=100)
plt.colorbar(viewing_card, label="normalized intensity (I/I_0)")

# to title the viewing card
plt.title(" 2D double slit normalised intensity pattern ")  

# Set aspect ratio
plt.gca().set_aspect(3/2)

plt.xlabel(" x position on detector (mm)")                       # label the x axis
plt.ylabel(" y position on detector (mm)")                       # label the y axis

# limit the x and y axes for a better view of the plot
plt.xlim(-5,5)                                                   
plt.ylim(-5,5)                                               

plt.show()                                                       # to show the plot  
```
